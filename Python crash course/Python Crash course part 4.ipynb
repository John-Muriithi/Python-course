# Map() function

```python
seq = [1,2,3,4,5]
```

```python
def times2(var):
    return var*2
```

```python
list(map(times2,seq))
```

# Lambda expression or Anonymous function

```python
# you dont need to define the function times2
```

```python
answer= lambda var:var*2
```

```python
list(map(lambda num:num*2,seq))
```

# Filter() Function

Filters out elements from a sequence

```python
#Returning only even numbers
```

```python
list(filter(lambda num: num%2 == 0,seq))
```

# Tuple Unpacking

```python
x = [(1,2),(3,4),(5,6)]  # List of Tuples
```

```python
x[0][1]
```

```python
#But the same can be done using tuple unpacking
```

```python
for item in x:
    print(item)
```

```python
for a,b in x:
    print(a)
```

```python
for a,b in x:
    print(b)
```

```python
for a,b in x:
    print(a)
    print(b)
```
