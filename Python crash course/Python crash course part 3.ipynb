## While Loop

```python
i = 1

while i < 5:
    print('i is: {}'.format(i))
    i = i+1
```

List apprehension

```python
x = list(range(5))
```

```python
out = []

for num in x:
    out.append(num**2)
    
```

```python
print(out)
```

Now This for loop into List comprehension

```python
[num**2 for num in x]
```
