# Pandas

 Series : The first main data types we gonna be working on with pandas

```python
import numpy as np
```

```python
import pandas as pd
```

```python
# Lets have 4 python separate objects

labels = ['a','b','c']
my_data = [10,20,30]
arr = np.array(my_data)
d = {'a':10,'b':20,'c':30}
```

```python
# Now lets create a series

pd.Series(data = my_data)
# well it looks like numPy array but notice theres an index.
```

```python
#Notice we can specify what our index is

pd.Series(data= my_data,index = labels)
```

```python
#You dont have to repeat data and index

pd.Series(my_data,labels)
```

```python
#A numpy array or python list may both define the data

pd.Series(arr,labels)
```

```python
#what if you pass a dictionary? the keys become labels and items- data

pd.Series(d)
```

# Data points can be any type

```python
pd.Series(data = labels)
```

```python
pd.Series(data = [sum,print,len,max])
```
