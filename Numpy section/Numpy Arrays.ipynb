## Numpy Arrays

Can be 1D vectors or 2D Matrices

```python
my_list = [1,2,3]
```

```python
my_list
```

```python
import numpy as np
```

```python
#casting a normal python list into an array

np.array(my_list)
```

```python
arr = np.array(my_list)
```

```python
arr
```

```python
#creating a matrix using list of lists

my_mat = [[1,2,3],[4,5,6],[7,8,9]]
my_mat
```

```python
# This gives a 2D array

np.array(my_mat)
```

Numpy Built-in methods that create arrays faster and simpler

## arange() zeros() ones()

```python
np.arange(0,10)
```

```python
#Third argument is step size

np.arange(0,11,2)
```

```python
#1D Vector of zeros - pass a single digit

np.zeros(3)
```

```python
#2D Matrix of zeros - pass a tuple (row,column)

np.zeros((5,5))
```

```python
np.ones(4)
```

```python
np.ones((3,4))
```

## Linspace()

```python
np.linspace(0,5,10)
```

```python
np.linspace(0,5,100)
```

# Identity Matrix Using NumPy

```python
np.eye(4)
```

# Arrays of random numbers

```python
np.random.rand(5)   #-Returns 1D 
```

```python
np.random.rand(5,5)   # 2D REM: dont pass tuple
```

A sample from the normal distribution or gaussian distrubution we use randn()

```python
np.random.randn(2)  #1D
```

```python
np.random.randn(3,3)  #2D
```

Random integers from a high to low number

```python
#randint(low,high,size)

np.random.randint(1,100)
```

```python
np.random.randint(1,100,10) #inclusive of low end but exclusive of high end
```
