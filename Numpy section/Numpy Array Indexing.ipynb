```python
import numpy as np
```

```python
arr = np.arange(0,11)
```

```python
arr
```

```python
arr[8]
```

```python
#Slicing Notation

arr[1:5]
```

```python
arr[:6]
```

```python
arr[5:]
```

```python
arr[0:5] = 100
```

```python
arr
```

```python
#Reset

arr = np.arange(0,11)
arr
```

```python
slice_of_arr = arr[0:6] #Grab everything upto but not including index 6
```

```python
slice_of_arr
```

```python
slice_of_arr[:] = 99   #grab everything and equate to 99
```

```python
slice_of_arr
```

```python
#If i call the array now, it actually has changed, not just on the slice

arr
```

# Indexing a 2D Array

```python
arr_2d = np.array([[5,10,15],[20,25,30],[35,40,45]])
```

```python
arr_2d
```

Two methods of grabbing elements from 2d array
1. Double bracket format
2. Single bracket with comma

```python
#Double Bracket Format

arr_2d[0][0]   # [row][column]
```

```python
arr_2d[0] #returns the first row
```

```python
arr_2d[1][1]
```

```python
arr_2d[2][2]
```

```python
#Single Bracket Format

arr_2d[0,0]
```

```python
arr_2d[1,1]
```

```python
arr_2d[2,1]
```

# What if you wanted some matrices from this matrix

```python
arr_2d[:2,1:]
#grab everything upto but not including row 2 and from column 1 onwards
```

```python
arr_2d[:3,1:]
```

```python
arr_2d[:2,:2]
```

```python
arr_2d[:2]
```

```python
arr_2d[1:2]
```

```python
arr_2d[0:,1:2]
```

# Conditional selection

```python
arr = np.arange(1,11)
```

```python
arr
```

```python
bool_arr = arr > 5
```

```python
bool_arr
```

```python
arr[bool_arr]
```

```python
arr[arr>5]
```

```python
arr[arr<3]
```
